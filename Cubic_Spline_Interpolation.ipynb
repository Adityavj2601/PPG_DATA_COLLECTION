```
def function(data_something , time , dataframe_values):
  var1 = time
  var2 = dataframe_values
  #print(var1)
  #print(var2     )
  ans = interpolate.splrep(var1 , var2)

  #print(ans)
  return (interpolate.splev(data_something , ans))

def findValley(dataframe):
  vally = []
  array_16_val_vally = []
  array_16_loc_vally = []
  for i in range(1,len(dataframe)):
    array_16_val_vally.append(dataframe[i])
    array_16_loc_vally.append(i)
    
    if(i % 16== 0):
      mean = stc.mean(array_16_val_vally)
      standrad_deviation = stc.stdev(array_16_val_vally)
      for k in array_16_val_vally:
        z_score_var = ((k - mean) / standrad_deviation)
        if z_score_var < -1.1:
          value = int(array_16_val_vally.index(k))
          vally.append((array_16_loc_vally[value],k)) 
          
          
      array_16_loc_vally=[]
      array_16_val_vally=[]
  return vally


#Fetch Location and value of Data 
def fetch_Location_Value(data):
  loc = [lis[0] for lis in data]
  val = [lis[1] for lis in data]
  return loc , val 
```
